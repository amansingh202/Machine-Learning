# BootstrapAggregation (Bagging)
Method to resolve overfitting for classification or regression problems <br> It does so by taking the random samples
of the original data. <br> The number of samples must be of the same size as the original dataset with replacement
or with duplicates <br> A base model is trained on each bootstrap sample. Since,each model is trained on slightly different
subset of data due to bootstrapping <br> Thus, they will be different of each other <br> Once all the models are trained
predictions are made on each individual model. <br>
The final prediction is determined by averaging the predictions from all the models (for regression tasks) <br>
Or by using voting for classification tasks, where the common prediction is chosen


```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
```

```python
data = datasets.load_wine(as_frame=True)

X = data.data
y = data.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=22)
```

```python
dtree = DecisionTreeClassifier(random_state=22)
dtree.fit(X_train, y_train)
```
