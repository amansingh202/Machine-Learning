```python
#regression when linear equation isn't able to fit the data points

import matplotlib.pyplot as plt

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

plt.scatter(x,y)
plt.show()
```

```python
import numpy as np

#equation as per the data points

mymodel = np.poly1d(np.polyfit(x,y,3))

print(mymodel)
```

```python
myline = np.linspace(1,22,100)
print(myline)
```

```python
plt.scatter(x,y)
plt.plot(myline, mymodel(myline))
plt.show()
```

# R squared
Relationship between the x and y values
R values ranges from 0 to 1
0 means no relation
1 means 100% related

```python
from sklearn.metrics import r2_score

mymodel = np.poly1d(np.polyfit(x,y, 3))

print(r2_score(y, mymodel(x)))
```

```python
#predict the model 
print(mymodel(17))
```
