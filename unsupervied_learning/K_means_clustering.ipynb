# K-means Clustering

```python
import matplotlib.pyplot as plt

x = [1.2, 1.5, 1.8, 2.0, 2.3, 3.0, 3.2, 3.5, 3.8, 3.9, 5.0, 5.1, 5.3, 5.5, 5.8]
y = [4.0, 4.2, 4.5, 4.1, 3.8, 2.5, 2.8, 2.9, 3.2, 3.0, 1.0, 0.8, 0.9, 1.2, 1.1]

plt.scatter(x,y)
plt.plot()
```

```python
#find the number of clusters required
#this is the elbow method 
#here the point after which there is a linear decrease in the number ofthe clusters
#is the actual number of clusters which is going to be needed in our model
from sklearn.cluster import KMeans

data = list(zip(x,y))
inertias = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)
    
plt.plot(range(1,11), inertias, marker = 'o')
plt.title('elbow method')
plt.xlabel('number of clusters')
plt.ylabel('Inertia')
plt.show()
```
