# train / test 
to measure the accuracy of the model

training data 80%

test data 20%

```python

import numpy as np
import matplotlib.pyplot as plt
np.random.seed(2)

#mean 3 and std deviation 1 for 100 number of samples
x = np.random.normal(3, 1, 100)

#mean 150 and std deviation 40 for 100 number of samples
y = np.random.normal(150, 40, 100) / x

plt.scatter(x,y)
plt.show()
```

```python
#split data into train and test 80%percent for training and 20 % for the model testing

train_X = x[:80]
train_y = y[:80]

test_X = x[80:]
test_y = y[80:]
```

```python
plt.scatter(train_X, train_y)
plt.show()
```

```python
plt.scatter(test_X, test_y)
plt.show()
```

```python
#fit the training data in the model

mymodel = np.poly1d(np.polyfit(train_X, train_y, 12))

myline = np.linspace(0, 6, 100)

plt.scatter(x,y)

plt.plot(myline, mymodel(myline))
plt.show()
```
