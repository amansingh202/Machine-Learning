# Linear Regression
This term regression is used to when you try to find the relationship variables

```python
import matplotlib.pyplot as plt
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, _, _, _ = stats.linregress(x, y)
```

```python
def predicted_y(x):
    return slope * x + intercept
    

```

```python
myModel = list(map(predicted_y, x))
```

```python
plt.scatter(x,y)
plt.plot(x, myModel)
plt.show()
```

```python
#predicting the output for this model for a new input value which is not present in the dataset
y_prediction = predicted_y(10)
print(f"Predicted value is : {y_prediction}")
```

# Bad Fit in Linear regression
where data are quite far from the prediction line

```python
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, _, _, _ = stats.linregress(x,y)
```

```python
def prediction_func(x):
    return x * slope + intercept
```

```python
y_pred = list(map(prediction_func, x))
```

```python
plt.scatter(x,y)
plt.plot(x, y_pred)
plt.show()
```
