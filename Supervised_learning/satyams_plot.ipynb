```python
import matplotlib.pyplot as plt

# Define the timestamps of each key emergency phase in minutes
timestamps = [0, 5, 10, 12.5, 17.5]

#awareness level
awareness_levels = [0, 1, 2, 2.1, 3]  


phases = ['Detection', 'Acknowledgment', 'Response Initiation', 'Response Underway', 'Resolution']


plt.figure(figsize=(10, 5))
plt.plot(timestamps, awareness_levels, '-o')  

#name of phases
for i, phase in enumerate(phases):
    plt.text(timestamps[i], awareness_levels[i] + 0.1, phase, ha='center')  

#title and x, y labels
plt.title('Emergency Awareness Over Time')
plt.xlabel('Time (minutes)')
plt.ylabel('Awareness and Response Level')

# y axis labels
plt.yticks([0, 1, 2, 3], ['Unaware', 'Detected', 'Acknowledged', 'Response Initiated'])


plt.grid()


plt.show()
```
