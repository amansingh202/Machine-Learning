# Hierarchical Clustering (unsupervised Learning)
Bottom up approach<br>
shortest distance between them to create large cluster


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

x = [4, 5, 10, 4, 3, 11, 14 , 6, 10, 12]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21]

plt.scatter(x,y)
plt.show()
```

```python
data = list(zip(x, y))

linkage_data = linkage(data, method = 'ward', metric = 'euclidean')
dendrogram(linkage_data)

plt.show()
```

# Agglomerative Clustering

```python
from sklearn.cluster import AgglomerativeClustering

hierarchical_clustering = AgglomerativeClustering(n_clusters=2, affinity = 'euclidean', linkage = 'ward')

labels = hierarchical_clustering.fit_predict(data)

plt.scatter(x, y, c = labels)
plt.show()
```
