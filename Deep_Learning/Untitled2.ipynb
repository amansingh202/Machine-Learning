```python
import numpy as np
import matplotlib.pyplot as plt

# Define the ReLU function and its derivative
def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return np.where(x < 0, 0, 1)

# Create an array of x values
x = np.linspace(-2, 2, 100)
g = relu(x)
g_prime = relu_derivative(x)

# Plotting
plt.figure(figsize=(12, 6))

# Plot g(x)
plt.subplot(1, 2, 1)
plt.plot(x, g, label='g(x) = ReLU(x)', color='blue')
plt.title('ReLU Activation Function')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid()
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.legend()

# Plot g'(x)
plt.subplot(1, 2, 2)
plt.plot(x, g_prime, label="g'(x) = ReLU'(x)", color='red')
plt.title('Derivative of ReLU Activation Function')
plt.xlabel('x')
plt.ylabel("g'(x)")
plt.grid()
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Define the function f(x; )
def f(x, m1, b1, m2, b2):
    return relu(m1 * x + b1) - relu(m2 * x + b2)

# Parameters
m1, b1, m2, b2 = 1, -1, 1, -2

# Create an array of x values
x_f = np.linspace(-1, 3, 100)
f_values = f(x_f, m1, b1, m2, b2)

# Plotting f(x; )
plt.figure(figsize=(6, 6))
plt.plot(x_f, f_values, label='f(x; )', color='green')
plt.title('Function f(x; )')
plt.xlabel('x')
plt.ylabel('f(x; )')
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the polynomial feature map function
def polynomial_feature_map_(x, degree):
    return np.array([x**i for i in range(degree + 1)])

def polynomial_feature_map(x_array, degree):
    return np.array([polynomial_feature_map_(x_array[i], degree) for i in range(len(x_array))])

# Inputs
x = np.linspace(-1, 2, 100)  # Generate 100 points between -1 and 2
degree = 2                   # Polynomial degree 2 (quadratic)
w = np.array([1, -2, 1])     # Coefficients for 1, x, x^2

# Generate polynomial features
X_tilde = polynomial_feature_map(x, degree)

# Compute the output y using the weight vector w
y = X_tilde @ w

# Plot the resulting polynomial
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = 1 - 2x + x^2')
plt.grid(True)
plt.show()
```
